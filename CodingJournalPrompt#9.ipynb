```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# We want to generate 1000 random numbers between [0, 1]
random_numbers = np.random.uniform(0, 1, 1000)

# This will create our histogram with 100 bins
plt.hist(random_numbers, bins = 100, edgecolor = 'k', alpha = 0.7)

# Axes labeled
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.savefig('uniform_distribution_plot.pdf')

plt.show
```
