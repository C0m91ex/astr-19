```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# We want to generate 100 random numbers using a normal Gaussian distribution
mean = 5
sigma = 1
random_numbers = np.random.normal(mean, sigma, 1000)

# We want to create a histogram with 100 bins
plt.hist(random_numbers, bins = 100, edgecolor = 'k', alpha = 0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.savefig('random_distribution_plot.pdf')

plt.show()
```
