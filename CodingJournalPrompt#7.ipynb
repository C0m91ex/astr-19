```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create the array x ranging form [0, 1] in 100 steps
x = np.linspace(0, 1, 100)

# Defined the exponential function that returns exp(x)
def exponential_function(x):
    return np.exp(x)
```

```python
# Calculates the y = exp(x)
y = exponential_function(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Awesomeness')

# Axes labels
plt.xlabel('Time [milliseconds]')
plt.ylabel('Awesomeness')

plt.legend()

plt.savefig('awesomeness_plot.pdf')

plt.show()
```
