```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# We want to create the array for x that ranges from [0, 1]
x = np.linspace(0, 1, 100)

# sin function
def sin_function(x):
    return np.sin(x)

# cos function
def cos_function(x):
    return np.cos(x)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# plot the sin function
axes[0].plot(x, sin_function(x))
axes[0].set_title('sin(x)')  

axes[1].plot(x, cos_function(x))
axes[1].set_title('cos(x)')  

for ax in axes:
    ax.set_xlabel('x')
    ax.set_ylabel('y')

plt.tight_layout()

plt.savefig('sin_cos_plot.pdf')

plt.show()
```
